```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('covid-19-data/us-counties.csv')
df.head()
```

```python
kcMetro = df[
    (df['state'] == 'Missouri') & (df['county'] == 'Kansas City')
    | (df['state'] == 'Missouri') & (df['county'] == 'Jackson')
    | (df['state'] == 'Missouri') & (df['county'] == 'Clay')
    | (df['state'] == 'Kansas') & (df['county'] == 'Wyandotte')
    | (df['state'] == 'Kansas') & (df['county'] == 'Johnson')
    | (df['state'] == 'Kansas') & (df['county'] == 'Douglas')
]
kcMetro.head()
```

```python
fig_dims = (15, 8)
fig, ax = plt.subplots(figsize=fig_dims)
x_dates = kcMetro['date'].unique()
ax.set_xticklabels(labels=x_dates, rotation=45, ha='right')
ax.set(title='KC Counties')
sns.scatterplot(x='date', y='cases', hue='county', data=kcMetro, ax=ax)
```

```python
fig_dims = (15, 10)
fig, ax = plt.subplots(figsize=fig_dims)
x_dates = kcMetro['date'].unique()
ax.set_xticklabels(labels=x_dates, rotation=45, ha='right')
ax.set(yscale='log', ylim=(1, 500), title='KC Counties, Log Y Axis')
sns.scatterplot(x='date', y='cases', hue='county', data=kcMetro, ax=ax)
```
